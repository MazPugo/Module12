```python
import matplotlib.pyplot as plt
```

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]
```

```python
names = list(map(lambda tup:tup[0], salaries))
```

```python
salary_values = list(map(lambda tup:tup[1], salaries))
```

```python
plt.bar(names, salary_values)
```

```python
plt.bar(names, salary_values,color ='green')
```

```python
plt.barh(names, salary_values)
```

```python
plt.barh(names, salary_values,color ='green')
```
